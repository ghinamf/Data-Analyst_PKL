## 3. Data prep

### OLD DATASET

```python
#dropping unwanted columns [df2]
df2= df.drop(['Tahun Bulan', 'Issue ID', 'Tracker', 'State', 'Start Date', 'Due Date', 'Poin', 'Subject', 'ID Creator', 'Nama Creator', 'Ts Menit', 'St Lembur'], axis=1)
df2.tail(15)
```

```python
#checking the column of 'Sumber'
df2['Sumber']
```

```python
#just taking the Lapor PPDB rows [df3]
df3 = df2[df2['Sumber'] != "Lapor SIAP"]
df3
```

```python
#checking null values
df3.isnull().sum()
```

```python
#filling the null/nan calues with 'Missing'
df3.fillna('Missing', inplace=True)
```

```python
df3.dtypes
```

```python
#changing the types of data to datetime so it can be calculated
df3[['Created On', 'Closed On', 'Wkt Resolved']] = df3[['Created On', 'Closed On', 'Wkt Resolved']].apply(pd.to_datetime, format='mixed')
df3.dtypes
```

```python
#durations calculation in HOUR
df3['Durations'] = (df3['Wkt Resolved'] - df3['Created On']).dt.total_seconds() / 3600

#Float to integer conv
df3['Durations'] = df3['Durations'].astype(int)
df3['ID Resolver'] = df3['ID Resolver'].astype(int)
```

```python
df4= df3.drop(['Created On', 'Closed On', 'Wkt Resolved'], axis=1)
df4
```

```python
#calculating the total project based on the nunique of 'Nama Resolver' then combined 
aa=df4.groupby('Nama Resolver')['Project'].nunique()
df4['Total Project']= df4['Nama Resolver'].map(aa)
df4
```

```python
#calculating the total time in HOUR
totalTime_perResolver= df4.groupby('Nama Resolver')['Durations'].sum().astype(int)
df4['Total Time']= df4['Nama Resolver'].map(totalTime_perResolver)
```

```python
df4.drop(['Sumber', 'Project'], axis=1)
```

```python
#dropping the duplicates from 'Nama Resolver'
df5 = df4.drop_duplicates(subset=['Nama Resolver'], keep='first')
df5= df5.drop(['Sumber', 'Project'], axis=1)
df5
```

```python
#exporting the data frame to csv
df5.to_csv(r'C:\Users\Lenovo\Downloads\pkl_rough\cleanedData.csv')
```

## Data Correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
#import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
#from matplotlib.pyplot import figure

%matplotlib inline
#matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#checking missing data
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#objt(str) to category
data['Grup Resolver'] = data['Grup Resolver'].astype('category')
data['Nama Resolver'] = data['Nama Resolver'].astype('category')
data['Kualifikasi'] = data['Kualifikasi'].astype('category')
data['Kelas Keahlian'] = data['Kelas Keahlian'].astype('category')
```

```python
data.dtypes
```

```python
#category to float
data['Nama Resolver'] = pd.factorize(data['Nama Resolver'])[0].astype(float)
data['Grup Resolver'] = pd.factorize(data['Grup Resolver'])[0].astype(float)
data['Kualifikasi'] = pd.factorize(data['Kualifikasi'])[0].astype(float)
data['Kelas Keahlian'] = pd.factorize(data['Kelas Keahlian'])[0].astype(float)
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Map')
plt.show()

#cor = data.corr()
#plt.figure(figsize = (70,70))
#sns.heatmap(cor, annot = True)
#plt.show()
```
