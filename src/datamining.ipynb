#### Encoding

LabelEncoder and OneHotEncoder
1. LabelEncoder
For example, if we use LabelEncoder to encode the colors ‘red’, ‘green’, and ‘blue’, the resulting numerical labels might be 0, 1, and 2, but there is no inherent ranking or ordering between these values.
2. OneHotEncoder
For example, if we use OneHotEncoder to encode the same colors ‘red’, ‘green’, and ‘blue’, the resulting matrix would have 3 columns, with a 1 in the column corresponding to the color of the sample and 0 in the other columns.
